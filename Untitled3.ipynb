```python
# Temperature Monitoring and Control System

def temperature_control():
    print("=== Temperature Monitoring System ===")

    # Set desired temperature range
    min_temp = 20
    max_temp = 30

    # Take temperature input
    current_temp = float(input("Enter current temperature (°C): "))

    print("Current Temperature:", current_temp, "°C")

    if current_temp > max_temp:
        print("Temperature is HIGH!")
        print("Cooling System (Fan) is ON ❄️")

    elif current_temp < min_temp:
        print("Temperature is LOW!")
        print("Heating System is ON 🔥")

    else:
        print("Temperature is NORMAL 😊")
        print("All Systems are OFF")

# Run the system
temperature_control()
```
